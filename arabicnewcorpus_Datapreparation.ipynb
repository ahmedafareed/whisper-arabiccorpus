```python
import csv
import json
import os


csv_file_path = '/content/wavtext.csv'
output_json_path = '/content/output.json'
wav_data_path = '/content/wav_data/'

dataset = []

with open(csv_file_path, 'r', encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file)

    for row in reader:
        if len(row) < 2:
            continue
        wav_file = row[0].strip()
        text = row[1].strip()
        folder_name = wav_file.split('-')[0]
        file_path = os.path.join(wav_data_path, folder_name, wav_file)
        dataset.append({
            "audio": file_path,
            "text": text
        })

# Write dataset to JSON file
with open(output_json_path, 'w', encoding='utf-8') as json_file:
    json.dump(dataset, json_file, ensure_ascii=False, indent=4)
print(f"JSON file saved at {output_json_path}")
```

```python
import json

json_file_path = '/content/output.json'
wav_list_file_path = '/content/drive/MyDrive/wav_file_list.txt'
cleaned_json_file_path = '/content/drive/MyDrive/cleaned_json_file.json'

with open(wav_list_file_path, 'r') as f:
    wav_file_paths = set(line.strip() for line in f)

with open(json_file_path, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

valid_entries = []

for entry in json_data:
    audio_path = entry.get("audio", "").strip()

    if audio_path in wav_file_paths:
        valid_entries.append(entry)
    else:
        print(f"Missing audio file: {audio_path} - Entry will be removed.")

# Save only valid entries to a new file
with open(cleaned_json_file_path, 'w', encoding='utf-8') as f:
    json.dump(valid_entries, f, ensure_ascii=False, indent=4)

print(f"Cleaned JSON file saved to {cleaned_json_file_path}")
```
